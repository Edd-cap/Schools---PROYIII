```python
def rename_columns(data):
    data = data.rename(columns={
        'Matriculados': 'num_matriculados',
        'Presentados': 'num_presentados',
        'Aptos': 'num_aptos',
        '% Aptos': 'porcentaje_aptos',
        'Media expediente': 'media_expediente',
        'Desviación típica': 'desviacion_tipica_expediente',
        'Media fase obligatoria': 'media_fase_obligatoria',
        'Desviación típica media PAU': 'desviacion_tipica_pau',
        'Diferencia absoluta entre media expediente y media PAU': 'dif_abs_media_expediente_pau'
    })
    return data

df = rename_columns(df)
```

```python
def calcular_rendimiento(data):
    # Pesos para las variables
    w_1 = w_2 = w_3 = 0.3
    w_4 = 0.1
    
    # Cálculo del rendimiento según la fórmula
    rendimiento = w_1 * data['porcentaje_aptos'] + \
                  w_2 * data['media_expediente'] + \
                  w_3 * data['media_fase_obligatoria'] - \
                  w_4 * data['dif_abs_media_expediente_pau']
    
    # Normalización min-max scaling
    rendimiento_normalizado = (rendimiento - rendimiento.min()) / (rendimiento.max() - rendimiento.min())
    
    return rendimiento_normalizado
# Reemplazar los valores faltantes en la columna de rendimiento por 0

# Ejemplo de uso
df['rendimiento'] = calcular_rendimiento(df)
```

```python
# Reemplazar los valores faltantes en la columna de rendimiento por 0
df['rendimiento'].fillna(0, inplace=True)
```

```python
def calcular_valor_instalaciones(data):
    # Lista de columnas de servicios e instalaciones
    columnas_servicios = ['Comedor', 'Instalaciones deportivas', 'Transporte', 'Cafetería',
                          'Inst + Gimnasio', 'Salón de actos', 'Pilota', 'Cursos de verano',
                          'Huerto escolar', 'Aula de idiomas', 'Inst + Gimnas + Piscina',
                          'Piscina', 'Gimnas + Piscina', 'Castellano', 'Valenciano', 'Francés',
                          'Alemán', 'Italiano']
    
    # Pesos para servicios menos comunes
    pesos = {
        'Pilota': 2,
        'Cursos de verano': 2,
        'Piscina': 2,
        'Gimnas + Piscina': 2,
        'Alemán': 2,
        'Italiano': 2,
        'Huerto escolar': 2
    }
    
    # Calcular el valor de las instalaciones
    valor_instalaciones = data[columnas_servicios].sum(axis=1)  # Suma de servicios e instalaciones por colegio
    
    # Normalización Min-Max Scaling
    valor_instalaciones_normalizado = (valor_instalaciones - valor_instalaciones.min()) / (valor_instalaciones.max() - valor_instalaciones.min())
    
    return valor_instalaciones_normalizado

# Ejemplo de uso
df["valor_instalaciones"] = calcular_valor_instalaciones(df)
```

```python
def normalizar_variables(data):
    # Normalizar las variables dividiendo cada una por su máximo valor
    data_norm = data.copy()
    for column in ['zonas_verdes', 'densidad_poblacion', 'turismos_e', 'paro',
                   'renta_media', 'riesgo_pobreza', 'bus_count', 'metro_count']:
        data_norm[column] = data[column] / data[column].max()
    return data_norm

# Ejemplo de uso
df = normalizar_variables(df)
```

```python
def calcular_valor_barrio(data):
    # Pesos para las características del barrio
    pesos = {
        'zonas_verdes': 0.2,
        'densidad_poblacion': -0.1,
        'turismos_e': 0.1,
        'paro': -0.3,
        'renta_media': 0.4,
        'riesgo_pobreza': -0.2,
        'bus_count': 0.1,
        'metro_count': 0.1
    }
    
    # Calcular el valor del barrio según la fórmula
    valor_barrio = sum(data[column] * peso for column, peso in pesos.items())
    
    # Normalización Min-Max Scaling
    valor_barrio_normalizado = (valor_barrio - valor_barrio.min()) / (valor_barrio.max() - valor_barrio.min())
    
    return valor_barrio_normalizado

# Ejemplo de uso
df["valor_barrio"] = calcular_valor_barrio(df[['zonas_verdes', 'densidad_poblacion', 'turismos_e', 'paro',
                                                'renta_media', 'riesgo_pobreza', 'bus_count', 'metro_count']])
```

```python
def calcular_puntuacion_final(data):
    # Pesos para las columnas
    peso_rendimiento = 0.5
    peso_instalaciones = 0.3
    peso_barrio = 0.2
    
    # Calcular la puntuación ponderada para cada centro educativo
    data['puntuacion'] = (peso_rendimiento * data['rendimiento']) + \
                         (peso_instalaciones * data['valor_instalaciones']) + \
                         (peso_barrio * data['valor_barrio'])
    
    # Aplicar penalización si el centro educativo es de tipo "Privado"
    data.loc[data['Tipo'] == 'Privado', 'puntuacion'] -= 0.5
    
    # Normalización a escala de 0 a 10
    puntuacion_normalizada = (data['puntuacion'] - data['puntuacion'].min()) / (data['puntuacion'].max() - data['puntuacion'].min()) * 10
    
    return puntuacion_normalizada

# Ejemplo de uso
df['puntuacion_final'] = calcular_puntuacion_final(df)
```

```python
# Crear un nuevo DataFrame con las columnas 'Nombre' y 'puntuacion_final'
nuevo_df = df[['Nombre', 'puntuacion_final']]

# Ordenar el nuevo DataFrame por la columna 'puntuacion_final' de mayor a menor
nuevo_df = nuevo_df.sort_values(by='puntuacion_final', ascending=False)

# Mostrar el nuevo DataFrame ordenado
for i,a in enumerate(nuevo_df["Nombre"]):
    print(i,a)
```
