```python
# Grupo 1: Infraestructura y recursos
infraestructura_recursos = ['Biblioteca', 'Laboratorio / taller', 'Aula de informática', 'Aula de música', 
                            'Instalaciones deportivas', 'Transporte', 'Cafetería', 'Inst + Gimnasio', 'Salón de actos',
                            'Pilota', 'Cursos de verano', 'Huerto escolar', 'Aula de idiomas', 'Inst + Gimnas + Piscina',
                            'Piscina', 'Gimnas + Piscina']

# Grupo 2: Idiomas
idiomas = ['Castellano', 'Valenciano', 'Inglés', 'Francés', 'Alemán']

# Grupo 3: Niveles educativos ofrecidos
niveles_educativos = ['EI', 'BACH', 'FP', 'SEC']

# Grupo 4: Entorno y ubicación
entorno_ubicacion = ['zonas_verdes', 'densidad_poblacion', 'turismos_e', 'paro', 'renta_media']

# Grupo 5: Acceso y transporte
acceso_transporte = ['bus_count', 'metro_count']

# Grupo 6: Opiniones y reseñas
opiniones_reseñas = ['NotaReviews', 'NumReviews']



# Grupo 7: Actividades extracurriculares
actividades_extracurriculares = ['Futbol', 'Baloncesto', 'Baile', 'Ajedrez', 'Patinaje',
                                 'Guitarra', 'Judo', 'Teatro', 'Manualidades', 'Programacion', 'Becas', 'Excursiones_Convivencias', 'Enfermeria', 
                                 'PatioInfantil', 'CentroTecnologico', 'GoogleClassroom', 'Teams', 'HerramientasPropiasTecnologicas',
                                 'NecesidadesEspecialesAlumnos']
#Grupo 8
extras = ['Precio', 'Horario']

# Creamos un diccionario para almacenar los grupos
grupos_variables = {
    'Infraestructura y recursos': infraestructura_recursos,
    'Idiomas': idiomas,
    'Niveles educativos ofrecidos': niveles_educativos,
    'Entorno y ubicación': entorno_ubicacion,
    'Acceso y transporte': acceso_transporte,
    'Opiniones y reseñas': opiniones_reseñas,
    'Actividades extracurriculares': actividades_extracurriculares,
    'Extras': extras,
}

# Ahora puedes acceder a cada grupo de variables utilizando el nombre del grupo como clave en el diccionario.
# Por ejemplo:
print("Variables del grupo 'Infraestructura y recursos':", grupos_variables['Infraestructura y recursos'])
```

### Grupo 1

```python
def calcular_puntuacion_infraestructura_recursos(datos):
    infraestructura_recursos = ['Biblioteca', 'Laboratorio / taller', 'Aula de informática', 'Aula de música', 
                                'Instalaciones deportivas', 'Transporte', 'Cafetería', 'Inst + Gimnasio', 'Salón de actos',
                                'Pilota', 'Cursos de verano', 'Huerto escolar', 'Aula de idiomas', 'Inst + Gimnas + Piscina',
                                'Piscina', 'Gimnas + Piscina']
    variables_con_peso_mayor = ['Transporte', 'Pilota', 'Cursos de verano', 'Huerto escolar',
                                'Inst + Gimnas + Piscina', 'Piscina', 'Gimnas + Piscina']
    total_variables = len(infraestructura_recursos)
    suma_variables = sum(datos[var] * (2 if var in variables_con_peso_mayor else 1) for var in infraestructura_recursos)
    puntaje_grupo = suma_variables / total_variables
    return puntaje_grupo

# Calcular el puntaje del grupo 1 para cada fila del DataFrame y agregarlo como una nueva columna
df['Recursos'] = df.apply(calcular_puntuacion_infraestructura_recursos, axis=1)
```

### Grupo 2

```python
def calcular_puntuacion_idiomas(datos):
    # Variables del grupo 2: Idiomas
    idiomas = ['Castellano', 'Valenciano', 'Inglés', 'Francés', 'Alemán']
    
    # Calcular la puntuación del grupo
    total_idiomas = len(idiomas)
    suma_idiomas = sum(datos[idioma] for idioma in idiomas)
    puntaje_grupo = suma_idiomas / total_idiomas
    
    return puntaje_grupo

# Calcular el puntaje del grupo 2 para cada fila del DataFrame y agregarlo como una nueva columna
df['idiomas'] = df.apply(calcular_puntuacion_idiomas, axis=1)
```

### Grupo 3

```python
def calcular_puntuacion_nivel_educativo(datos):
    # Variables del grupo 3: Niveles educativos ofrecidos
    niveles_educativos = ['EI', 'BACH', 'FP', 'SEC']
    
    # Calcular la puntuación del grupo
    total_niveles = len(niveles_educativos)
    suma_niveles = sum(datos[nivel] for nivel in niveles_educativos)
    puntaje_grupo = suma_niveles / total_niveles
    
    return puntaje_grupo

# Calcular el puntaje del grupo 3 para cada fila del DataFrame y agregarlo como una nueva columna
df['nivel_educativo'] = df.apply(calcular_puntuacion_nivel_educativo, axis=1)
```

### Grupo 4

```python
def calcular_puntuacion_entorno_ubicacion(datos):
    # Variables del grupo 4: Entorno y ubicación
    entorno_ubicacion = ['zonas_verdes', 'densidad_poblacion', 'turismos_e', 'paro', 'renta_media']
    
    # Seleccionar los datos relevantes
    datos_entorno = datos[entorno_ubicacion]
    
    # Redimensionar los datos a 2D si es necesario
    if len(datos_entorno.shape) == 1:
        datos_entorno = datos_entorno.values.reshape(-1, 1)
    
    # Normalizar los datos
    scaler = MinMaxScaler()
    datos_entorno_normalizados = scaler.fit_transform(datos_entorno)
    
    # Calcular la puntuación del grupo
    puntaje_grupo = datos_entorno_normalizados.mean()
    
    return puntaje_grupo

# Calcular el puntaje del grupo 4 para cada fila del DataFrame y agregarlo como una nueva columna
df['entorno_ubicacion'] = df.apply(calcular_puntuacion_entorno_ubicacion, axis=1)
```

### Grupo 5

```python
def calcular_puntuacion_acceso_transporte(datos):
    # Variables del grupo 5: Acceso y transporte
    acceso_transporte = ['bus_count', 'metro_count']
    
    # Calcular la puntuación del grupo
    puntaje_grupo = datos[acceso_transporte].mean()
    
    return puntaje_grupo

# Calcular el puntaje del grupo 5 para cada fila del DataFrame y agregarlo como una nueva columna
df['acceso_transporte'] = df.apply(calcular_puntuacion_acceso_transporte, axis=1)

df['acceso_transporte'] = df['acceso_transporte'].div(df['acceso_transporte'].max())
```

### Grupo 6

```python
def calcular_puntuacion_opiniones_reseñas(datos):
    # Variables del grupo 6: Opiniones y reseñas
    nota_reviews = datos['NotaReviews']
    num_reviews = pd.Series(datos['NumReviews'])
    num_reviews_max = num_reviews.max()

    
    nota_reviews_normalizada = (nota_reviews - 1) / 4  # Normaliza la nota de 1-5 a 0-1
    num_reviews_normalizado = num_reviews / num_reviews_max
    # Calcula la puntuación del grupo
    puntaje_grupo = nota_reviews_normalizada * 0.7 + num_reviews_normalizado * 0.3
    
    return puntaje_grupo

# Calcular el puntaje del grupo 6 para cada fila del DataFrame y agregarlo como una nueva columna
df['opiniones_reseñas'] = df.apply(calcular_puntuacion_opiniones_reseñas, axis=1)
```

### Grupo 7

```python
def calcular_puntuacion_actividades_extracurriculares(datos):
    # Variables del grupo 7: Actividades extracurriculares
    actividades_extracurriculares = datos[['Futbol', 'Baloncesto', 'Baile', 'Ajedrez', 'Patinaje',
                                 'Guitarra', 'Judo', 'Teatro', 'Manualidades', 'Programacion', 'Becas', 'Excursiones_Convivencias', 'Enfermeria', 
                                 'PatioInfantil', 'CentroTecnologico', 'GoogleClassroom', 'Teams', 'HerramientasPropiasTecnologicas',
                                 'NecesidadesEspecialesAlumnos']]
    
    # Calcular la puntuación del grupo
    puntaje_grupo = actividades_extracurriculares.sum() / len(actividades_extracurriculares)
    
    return puntaje_grupo

# Calcular el puntaje del grupo 7 para cada fila del DataFrame y agregarlo como una nueva columna
df['actividades_extracurriculares'] = df.apply(calcular_puntuacion_actividades_extracurriculares, axis=1)
```

### Grupo 8

```python
precio = df['Precio'].astype(str)
horario = df['Horario'].astype(str)
```

```python
def calcular_puntuacion_precio_horario(datos):
    # Variables del grupo 8: Precio y horario
    precio = datos['Precio']
    horario = datos['Horario']
    
    # Inicializar la puntuación
    puntaje_grupo = 0
    
    # Asignar puntajes según las condiciones
    if (precio == "Gratuito o <100€" and horario == "Horario ampliado de mañana y tarde"):
        puntaje_grupo = 1
    elif (precio == "Gratuito o <100€" or horario == "Horario ampliado de mañana y tarde"):
        puntaje_grupo = 0.5
    
    return puntaje_grupo

# Calcular el puntaje del grupo 8 para cada fila del DataFrame y agregarlo como una nueva columna
df['precio_horario'] = df.apply(calcular_puntuacion_precio_horario, axis=1)
```

```python
def calcular_nota_total(datos):
    # Pesos para cada grupo
    pesos_grupos = {
        'Recursos': 0.3,
        'idiomas': 0.15,
        'nivel_educativo': 0.05,
        'entorno_ubicacion': 0.2,
        'acceso_transporte': 0.1,
        'opiniones_reseñas': 0.05,
        'actividades_extracurriculares': 0.1,
        'precio_horario': 0.05
    }
    
    # Calcular la nota total
    nota_total = sum(datos[group] * peso for group, peso in pesos_grupos.items())
    
    # Escalar la nota total de 0 a 10
    nota_total_escalada = nota_total * 10
    
    return nota_total_escalada 

df['nota_total'] = df.apply(calcular_nota_total, axis=1)
df_ordenado = df.sort_values(by='nota_total')

print(df_ordenado['Nombre'])

```
